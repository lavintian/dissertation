```python
import pandas as pd
import numpy as np
```

```python
#load cleaned text file
dftest = pd.read_csv('/project/CLEAN_sentences.csv')
```

```python
from ast import literal_eval
```

```python
dftest['clean4']= dftest['clean3'].apply(literal_eval)
```

```python
from sentence_transformers import SentenceTransformer
```

```python
output_table_fin = pd.DataFrame()
model = SentenceTransformer('roberta-large-nli-stsb-mean-tokens')
#ID= 1000
for sentences in dftest['sentence'].tolist():

    embeddings = model.encode(sentences)
    output_table_1 =pd.DataFrame(embeddings).T
    output_table_fin = pd.concat([output_table_fin,output_table_1],axis=0)
```
